### **KEY AXIOMS AND DEFINITIONS**

**Axiom 1.** Every active voice in a sonority contributes to the listener's perception of musical structure.

**Definition 1.** For standardization, let the total sound of any musical entity (note, chord, etc.) be 1. 


### **METHODOLOGY**

When considering a musical entity, one must consider the parameters in which we can determine how loud/soft one note in the entity is in relation to other notes inside this entity.

Hence, the components of any musical entity can be represented by a vector with components that represent the $n^{th}$ note's relative balance.

Let the Voice Balancing Vector (VBV) be denoted with the vector $\vec{b}$, where $b_1, b_2, b_3 ...  b_n$ are the $n$ notes in the musical entity from top down.

Then:
$$
\vec{b} = \begin{bmatrix} 
                b_1 \\
                b_2 \\
                b_3 \\
                \vdots \\
                b_n
            \end{bmatrix}
$$

### **Beethoven Sonata Op. 111, First Movement Example**

Using this method, I will analyze the first movement of Beethoven's Sonata Op. 111 using Python's music21.

```python
from music21 import converter

score = converter.parse("data/beethoven_111.mxl")

for i, part in enumerate(score.parts):
    print(i, part.partName)

print(len(score.parts[0].getElementsByClass("Measure")))

measure = score.parts[1].measure(0)
measure.show("text")
```

```python
import pretty_midi

midi = pretty_midi.PrettyMIDI("data/beethoven111.mid")

notes = sorted(
    midi.instruments[0].notes,
    key=lambda note: note.start
)

for note in notes:
    print(
        pretty_midi.note_number_to_name(note.pitch),
        note.velocity,
        note.start,
        note.end
    )
```
